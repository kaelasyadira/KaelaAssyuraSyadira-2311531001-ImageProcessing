```python
import cv2
img = cv2.imread('merak.jpg')
```

```python
import cv2
from google.colab.patches import cv2_imshow

(h, w) = img.shape[:2]

center = (w // 2, h // 2)

matrix_45_ccw = cv2.getRotationMatrix2D(center, 45, 1.0)

rotated_45_ccw_img = cv2.warpAffine(img, matrix_45_ccw, (w, h))

print("Foto asli:")
cv2_imshow(img)
print("\n")
print("Foto dengan Rotasi 45 Derajat:")
cv2_imshow(rotated_45_ccw_img)
```

```python
import cv2
from google.colab.patches import cv2_imshow

matrix_180_ccw = cv2.getRotationMatrix2D(center, 180, 1.0)

rotated_180_ccw_img = cv2.warpAffine(img, matrix_180_ccw, (w, h))

print("Foto dengan Rotasi 180 Derajat:")
cv2_imshow(rotated_180_ccw_img)
```
